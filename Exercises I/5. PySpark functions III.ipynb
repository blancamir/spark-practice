# PySpark Collect()


It is used to retrieve the action output when you have very small result set

Difference: select() is a transformation function whereas collect() is an action.


Create a Dataframe

```python
import findspark
findspark.init()

import pyspark
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('sparkexamples').getOrCreate()
```

```python
spark
```

```python
airport = spark.read.csv('C:/Users/usuario\Documents\Blanca\SPARK\practica_pyspark/airport_codes.csv', inferSchema=True, header=True)
airport.show(3)
```

```python
airport.printSchema()
```

show() function on DataFrame prints the result of DataFrame in a table format

collect() retrieves all elements in a DataFrame as an array to the driver. printing a resultant array yields below output.

```python
airportCollect = airport.collect()
print(airportCollect)
```

```python
for row in airportCollect:
    print(row['name'] + "," + str(row['elevation_ft']))
```

```python
airportCollect2 = airport.select('name').collect()
print(airportCollect2)
```
